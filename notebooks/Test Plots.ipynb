# NN results

```python
PATH='/home/anowak/projects/deep-kernel-transfer/save/regression_multisines_nncheckpoints/sines/results_sines.npy'
```

```python
data = np.load(PATH, allow_pickle=True).item()
```

```python
idx = np.argsort(data["x_list"][0],axis=0).reshape(-1)
```

```python
def plot_for_n(data,n=0):
    idx = np.argsort(data["x_list"][n],axis=0).reshape(-1)
    fig, (ax, ax2) = plt.subplots(1,2, figsize=(8*2,3*2))
    ax.plot(data["x_list"][n][idx], data["y_list"][n][idx, 0], linestyle="--", color="green")
    ax.plot(data["x_list"][n][idx], data["y_list"][n][idx, 0], linestyle='', marker='o', color="green")
    ax.plot(data["x_list"][n][idx], data["mean"][n][idx, 0])
    ax.fill_between(data["x_list"][n][idx].reshape(-1), data["lower"][n][idx, 0], data["upper"][n][idx, 0], alpha=0.2)

    ax2.plot(data["x_list"][n][idx], data["y_list"][n][idx, 1], linestyle="--", color="green")
    ax2.plot(data["x_list"][n][idx], data["y_list"][n][idx, 1], linestyle='', marker='o', color="green")
    ax2.plot(data["x_list"][n][idx], data["mean"][n][idx, 1])
    ax2.fill_between(data["x_list"][n][idx].reshape(-1), data["lower"][n][idx, 1], data["upper"][n][idx, 1], alpha=0.2)
    plt.show()
```

```python
plot_for_n(data, 1)
```

```python
plot_for_n(data, 0)
```

```python
plot_for_n(data, 2)
```
